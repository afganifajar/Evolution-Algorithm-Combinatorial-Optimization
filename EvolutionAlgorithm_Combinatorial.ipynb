```
import numpy as np
import random
import pandas as pd
```

```
popSize = 5
maxIteration = 2

totalJob = 4
totalWorker = 7

crossover_rate = 0.8
mutation_rate = 0.2

crossover_total = int(popSize * crossover_rate)
mutation_total = int(popSize * mutation_rate)

costTable = [[3, 2, 4, 4, 4, 5, 6], [4, 5, 6, 4, 4, 4, 5], [4, 5, 2, 5, 3, 3, 3], [5, 4, 3, 5, 5, 3, 4]]
```

```
def createChromosome():
  chromosome = []
  gen = random.randint(1, totalWorker)
  while len(chromosome) < totalJob:
    if gen in chromosome:
      gen = random.randint(1, totalWorker)
    else:
      chromosome.append(gen)
  return chromosome


def countCost(chromosome):
  totalCost = 0
  for i in range(len(chromosome)):
    j = chromosome[i]-1
    cost = costTable[i][j]
    totalCost+=cost
  return totalCost


def addPopulation(chromosome, population):
  cost = countCost(chromosome)
  fitness = 100/cost
  individu = [chromosome, cost, fitness]
  population.append(individu)


def showPopulation():
  df = pd.DataFrame(populasi)
  df.columns = ["Kromosom", "Biaya", "Fitness"]
  print(df)
  print()


def initialization():
  population = []
  for i in range(popSize):
    chromosome = createChromosome()
    addPopulation(chromosome, population)
    
  print("[Populasi]:")
  showPopulation()
  return population


def crossover(population):
  print("[Crossover]:")
  cutPoint = random.randrange(1, totalJob)
  print("cut-point = " + str(cutPoint))
  print()

  for i in range(crossover_total):
    child = []

    index1 = random.randrange(popSize)
    index2 = random.randrange(popSize)
    while index2 == index1:
      index2 = random.randrange(popSize)

    parent1 = population[index1][0]
    parent2 = population[index2][0]
    
    for j in range(cutPoint):
      child.append(parent1[j])
    
    k = 0
    for j in range(cutPoint, totalJob):
      while parent2[k] in child:
        k+=1
      else:
        child.append(parent2[k])
    
    addPopulation(child, population)
    print("Crossover " + str(i+1))
    print("Parent 1 = " + str(parent1))
    print("Parent 2 = " + str(parent2))
    print("Child    = " + str(child))
    print()
  print("Hasil Crossover:")
  showPopulation()


def mutation(population):
  print("[Mutasi]:")
  xp1 = random.randrange(totalJob)
  xp2 = random.randrange(totalJob)
  while xp2 == xp1:
    xp2 = random.randrange(totalJob)
  print("Exchange Point = " + str(xp1+1) + " & " + str(xp2+1))
  print()

  for i in range(mutation_total):
    parent = population[random.randrange(popSize)][0]
    child = parent[:]

    temp = child[xp1]
    child[xp1] = child[xp2]
    child[xp2] = temp

    addPopulation(child, population)

    print("Mutation " + str(i+1))
    print("Parent = " + str(parent))
    print("Child  = " + str(child))
    print()

  print("Hasil Mutasi:")
  showPopulation()


def selection(population):
  print("[Seleksi]:")
  population.sort(reverse=True, key=lambda x: x[2])

  print("Hasil Sorting:")
  showPopulation()

  del population[-popSize:]
  
  print("Hasil Seleksi:")
  showPopulation()
```

```
for i in range(maxIteration):
  print('ITERASI : ' + str(i+1))
  if i == 0:
    populasi = initialization()
  crossover(populasi)
  mutation(populasi)
  selection(populasi)
```
